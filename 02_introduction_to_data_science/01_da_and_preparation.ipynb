# Data Warehouse
- teljes vállalat adatait tartalmazza
- struktúrált adatok
    - relációs jellegű adatok, előre definiált séma, tipikusan: csillag vagy hópehely séma 
- több üzleti terület adatait is tartalmazza
- nagy mennyiségű, historikus adatok
- riportolásra és vizualizációra optimalizált
- bonyolult ETL/ELT folyamatok

# Data Marts
- egy adott üzleti területre fókuszál
- kevesebb adat
- egyszerűbb struktúra
    - sokszor üzleti kérdésre optimalizált, emiatt gyakran letisztultabb, egyszerűbb
    - de alapvetően ugyanazokat asémákat használja, mint a Data Warehouse
- operatívabb elemzésekhez is használható

# Data Lake
- nyers adattároló, ahová mindenféla adat bekerül
    - struktúrált: SQL táblák, CSV
    - félstruktúrált: JSON, XML
    - struktúrálatlan: képek, videók, logok
- inkább data science, ML felhasználás

# Big Data
Olyan nagy mennyiségű, gyorsan keletkező és változatos adatok összessége, amelyek kezelése a hagyományos adatfeldolgozó eszközökkel nem hatékony, ezért elosztott, párhuzamosított tárolást és feldolgozást igénylő technológiák alkalmazását teszi szükségessé.

- elosztott tárolás, pl.: object storage, tehát felhőalapú tárolás. Mint pl.: AWS S3, Azure Data Lake / Blob Storage, Google Cloud Service
- párhuzamos feldolgozás, pl.: Apache Spark

#### Ismérvek:
- Value - Jelentős értéket hozdoz(hat)
- Volume - Nagy adatmennyiség
- Velocity - Az adatok gyorsan generálódnak
- Variety - Struktúrált, félig struktúrált, struktúrálatlan
- Veracity - Az adatok minősége nem feltétlenül tökéletes
- Variability - Az adat üzleti jelentése is változhat

# Mesterséges Intelligencia
Artifical Intelligence (AI):
Olyan rendszerek összessége, amelyek emberi intelligenciát utánzó feladatokat végeznek.
Ez egy gyűjtőfogalom, nem egy konkrét módszer.
- chatbot
- arcfelismerés

Machine Learning (ML):
Az AI egy részhalmaza, ahol a rendszer adatokból tanul, nem előre programozott szabályok alapján működik.

Lényege:
- van tanító adat
- a modell mintákat tanul
- képes előrejelzést hozni

Példák:
- hitelbírálat
- termékajánlás

Deep Learning:
Az ML egy speciális típusa, amely többrétegű neurális hálókat használ.

Jellemzők:
- nagyon sok adat kell
- nagy számítási kapacitás (GPU)

Példák:
- képfelismerés
- önvezető autók
- ChatGPT szerű modellek
