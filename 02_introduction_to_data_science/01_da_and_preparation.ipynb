# Data Warehouse

**Definíció:**  
A Data Warehouse a vállalat teljes adatvagyonát integrált formában tartalmazó,  
**elemzésre és riportolásra optimalizált** adattárház.

**Fő jellemzők:**
- A teljes vállalat adatait tartalmazza
- Strukturált adatok
  - Relációs jellegű adatok
  - Előre definiált séma
  - Tipikus modellek: **csillag** és **hópehely séma**
- Több üzleti terület (pl. értékesítés, pénzügy, marketing) adatait integrálja
- Nagy mennyiségű, historikus adatok
- Riportolásra és vizualizációra optimalizált
- Összetett **ETL / ELT** folyamatok

---

# Data Mart

**Definíció:**  
A Data Mart a Data Warehouse egy részhalmaza, amely egy adott üzleti terület igényeire fókuszál.

**Fő jellemzők:**
- Egy konkrét üzleti területet szolgál ki
- Kevesebb adat
- Egyszerűbb, átláthatóbb struktúra
  - Gyakran konkrét üzleti kérdésekre optimalizált
  - Alapvetően ugyanazokat a sémákat használja, mint a Data Warehouse
- Operatívabb, gyorsabb elemzésekre is alkalmas

---

# Data Lake

**Definíció:**  
A Data Lake egy központi, nyers adattároló, amelyben az adatok eredeti formájukban kerülnek eltárolásra.

**Adattípusok:**
- Strukturált: SQL táblák, CSV
- Félstrukturált: JSON, XML
- Strukturálatlan: képek, videók, logok

**Felhasználás:**
- Data Science
- Machine Learning
- Felfedező elemzések

# Big Data

**Definíció:**  
Olyan nagy mennyiségű, gyorsan keletkező és változatos adatok összessége,  
amelyek kezelése hagyományos eszközökkel nem hatékony, ezért  
**elosztott és párhuzamos feldolgozást igényel**.

**Technológiai jellemzők:**
- Elosztott tárolás
  - Objektumtárolók (felhő)
  - Példák: AWS S3, Azure Data Lake / Blob Storage, Google Cloud Storage
- Párhuzamos feldolgozás
  - Példa: Apache Spark

**Big Data ismérvek (6V):**
- **Value** – üzleti érték
- **Volume** – nagy adatmennyiség
- **Velocity** – gyors adatkeletkezés
- **Variety** – változatos adattípusok
- **Veracity** – adatminőség bizonytalansága
- **Variability** – üzleti jelentés változása

# Mesterséges Intelligencia

## Artificial Intelligence (AI)

**Definíció:**  
Az AI olyan rendszerek összessége, amelyek emberi intelligenciát utánzó feladatokat végeznek.

**Jellemzők:**
- Gyűjtőfogalom
- Nem egy konkrét algoritmus

**Példák:**
- Chatbotok
- Arcfelismerés

---

## Machine Learning (ML)

**Definíció:**  
Az AI egy részterülete, ahol a rendszer adatokból tanul,  
nem előre programozott szabályok alapján működik.

**Lényege:**
- Van tanító adat
- A modell mintázatokat tanul
- Képes előrejelzést vagy döntést hozni

**Példák:**
- Hitelbírálat
- Termékajánló rendszerek

---

## Deep Learning

**Definíció:**  
A Machine Learning speciális ága, amely többrétegű neurális hálókat használ.

**Jellemzők:**
- Nagy mennyiségű adat szükséges
- Jelentős számítási kapacitás (GPU)

**Példák:**
- Képfelismerés
- Önjáró járművek
- ChatGPT-szerű modellek

---

## Neurális hálók – Alapok

A mesterséges neuron a biológiai neuron által inspirált,  
**egyszerűsített matematikai modell**.

**Felépítése:**
- Bemenetek: \( x_1, \dots, x_n \) súlyokkal: \( w_1, \dots, w_n \)
- Bias: \( x_0, w_0 \)
- Súlyozott összeg + nemlineáris aktivációs függvény
- Kimenet: \( y \)
- Egyes hálótípusoknál lokális memória (visszacsatolás, késleltetés)

# Üzleti elemzés lépései (CRISP-DM)

**CRISP-DM:** Cross Industry Standard Process for Data Mining

## 1. Üzleti célok megértése
- Probléma definiálása
- Erőforrások feltérképezése
- Döntéshozók azonosítása
- Hipotézisek megfogalmazása
- Potenciális adatforrások

## 2. Adatok megismerése
- Adatgyűjtés (ETL)
- Alapstatisztikák
- Adatvizualizáció
- Adatminőség vizsgálata

## 3. Adatok előkészítése
- Adatkiválasztás
- Adattisztítás
- Változótranszformáció
- Új változók létrehozása
- Adatintegráció

## 4. Modellezés
- Változók kiválasztása
- Modell / algoritmus kiválasztása
- Modell tanítása

## 5. Értékelés
- Teljesítménymérés
- Validáció
- Üzleti célokkal való összevetés
- Következtetések levonása

## 6. Implementáció
- Modell élesítése
- Üzembe helyezés
- Folyamatos frissítés és monitorozás
